WHAT TRANSFORMERS CAN DO

```python
from transformers import pipeline

#pipeline () function connects a model with its necessary preprocessing and postprocessing steps
#This allows direct input of text and output of intelligible text

# creating a classifier objects
classifier = pipeline("sentiment-analysis")

# The model is downloaded and cached after creating the classifier m=object

# calling the classifier onject
classifier(["I do not hate this course",
            "I do not love it either",
            "I hope am going to have great fun with it",
            "Lets buckle up for some fun time"])
```


## Steps involved when text is passed to a pipeline

1. The text is preprocessed into a format the model can understand.
2. The preprocessed inputs are passed to the model
3. The predictions of the model are post-processed, so you can make sense of them.



## Available pipelines

```python
#zero-shot classification
# it allows you to specify which labels to use for the classification, so you don’t have to rely on the labels of the pretrained model

from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels = ["education", "lifestyle", "Politics"]

)
```

```python
# Text generation using a default model for task at hand
from transformers import pipeline

generator = pipeline("text-generation")
generator("In this course we will learn how to  ")
```

```python
# model hub where to find models for tasks like below  https://huggingface.co/models
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course we will learn how to",
    # 30 words maximum in length
    max_length=30,
    #two sentences will be returned
    num_return_sequences=2,
)
```

```python
# MASK FILLING using distilbert/distilroberta
# to fill in blanks in a given text

from transformers import pipeline

unmasker = pipeline("fill-mask", model="distilbert/distilroberta-base")
# top_k displays the number possibilities desired similar to num_return)seq
unmasker("This course will teach us all about <mask> models", top_k=2)
```

```python
# It is better to use f strings and pass the argument. The advantage of using an f-string is that it is intuitive to understand
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

mask_fill = pipeline("fill-mask", model="bert-base-multilingual-cased")
mask_fill(f"This course is {tokenizer.mask_token} and I have loved learning it.", top_k=2)
```

```python
#without using the f'string
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-base-multilingual-cased')
# Replace '[mask]' with '[MASK]' in the input string
unmasker("This course will teach us all about [MASK] models", top_k =2)
```

```python
#NAMED ENTITY RECOGNITION NER
# Task where the model has to find which names correspond to named entities like persons, places, mountains, cars ...

from transformers import pipeline

# grouped_entities=True allows model to regroup together parts of the sentence that correspond to same entity
ner = pipeline("ner", grouped_entities=True)
ner("My name is Vonne and I work at Jamii Telecom in Nairobi")
```

## TRANSFORMERS AND HOW THEY WORK

This means they have been trained on large amounts of raw text in a self-supervised fashion. Self-supervised learning is a type of training in which the objective is automatically computed from the inputs of the model. That means that humans are not needed to label the data!

This type of model develops a statistical understanding of the language it has been trained on, but it’s not very useful for specific practical tasks. Because of this, the general pretrained model then goes through a process called transfer learning. During this process, the model is fine-tuned in a supervised way — that is, using human-annotated labels — on a given task.
